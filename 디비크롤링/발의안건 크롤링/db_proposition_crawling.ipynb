```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

page_num = 1 #발의안 날짜랑 제목 따는 페이지 순차 검색 위한 int => 페이지 변동 시에 이 값만 바뀜
bill_no = 2023703 #발의안 내용이랑 발의자 명단 뽑는 페이지 순차 검색 위한 int => 페이지 변동 시에 이 값만 바뀜
p_time = []
p_title = []
p_contents = []
p_proposers = []

for i in range(5):

    if (i % 100 == 0):
        url1 = urlopen("http://watch.peoplepower21.org/index.php?mid=Euian&show=1&page=" + str(page_num) + "&title=&rec_num=100&lname=&sangim=&bill_result=")
        bsObject1 = BeautifulSoup(url1, "html.parser")
        cnt = 0
        page_num += 1
        for temp in bsObject1.tbody.find_all('td'):
            if (cnt % 5 == 0):
                p_time.append(temp.text.strip()) # 발의 날짜
                cnt = cnt + 1
            elif (cnt % 5 == 1):
                p_title.append(temp.text.strip()) # 발의 
                cnt = cnt + 1
            else:
                cnt = cnt + 1
        
    url2 = urlopen("http://watch.peoplepower21.org/?mid=LawInfo&bill_no=" + str(bill_no))
    bsObject2 = BeautifulSoup(url2, "html.parser")
    
    p_contents.append(bsObject2.find('div', {"class":"textType02"}).get_text().strip()[15:])
    #print(bsObject2.find('div', {"class":"textType02"}).get_text().strip()[15:])
    p_proposers.append(bsObject2.find('div', {"class":"col-sm-8"}).get_text().replace("\t","").replace(" ", "").replace("\n", ""))
    
    bill_no = bill_no - 1

    
propositionList = []

for i in range(0,len(p_contents)):
    info_pro = {'date' : p_time[i], 'title' : p_title[i], 'contents' : p_contents[i], 'proposers' : p_proposers[i]}
    propositionList.append(info_pro)
    print(propositionList[i])

#print(propositionList[:10])

    
'''
print(bsObject2.find('div', {"class":"textType02"}).get_text().strip()[15:])
print(bsObject2.find('div', {"class":"col-sm-8"}).get_text())
'''

```
