```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

bill_no = 583707 #페이지 순차 검색 위한 int => 페이지 변동 시에 이 값만 바뀜
p_startTime = []
p_endTime = []
p_category = []
p_participantsCnt = []
p_title = []
p_contents = []



for i in range(5):
    url2 = urlopen("https://www1.president.go.kr/petitions/" + str(bill_no))
    bsObject2 = BeautifulSoup(url2, "html.parser")
    
    #p_contents.append(bsObject2.find('h3', {"class":"partitionsView_title"}).get_text().strip())
    #print(bsObject2.find('div', {"class":"petitionsView_left_pg"}))
    #print(bsObject2.find('h3', {"class":"petitionsView_title"}).get_text())
    p_title.append(bsObject2.find('h3', {"class":"petitionsView_title"}).get_text())
    
    #print(bsObject2.find('h2', {"class":"petitionsView_count"}).get_text())
    p_participantsCnt.append(bsObject2.find('h2', {"class":"petitionsView_count"}).get_text())
    
    temp = bsObject2.find('div', {"class":"petitionsView_info"}).find_all('li')
    
    _category = temp[0].get_text()[4:]
    #print(_category)
    p_category.append(_category)
    
    _startTime = temp[1].get_text()[4:]
    #print(_startTime)
    p_startTime.append(_startTime)
    
    _endTime = temp[2].get_text()[4:]
    #print(_endTime)
    p_endTime.append(_endTime)
    
    _contents = bsObject2.find('div', {"class":"View_write"})
    _contents = str(_contents)
    _contents = _contents[60:-7].strip().replace("<br/>","").replace("\n","").replace("\r","")
    
    p_contents.append(_contents)
    #print(_contents)
    
    bill_no = bill_no - 1


petitionList = []

for i in range(0,len(p_contents)):
    info_pet = {'title' : p_title[i], 'category' : p_category[i], 'participants_count : p_participantsCnt[i], startTime' : p_startTime[i], 'endTime' : p_endTime[i], 'contents' : p_contents[i]}
    petitionList.append(info_pet)
    print(petitionList[i])

#print(propositionList[:10])
```
